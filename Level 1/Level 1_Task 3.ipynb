# Task 3 : Price Range Distribution

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
data
```

```python
# Display first few rows
data.head()
```

#  The Distribution of Price Ranges Among The Restaurants.

```python
# Count restaurants in each price range
price_range_counts = data['Price range'].value_counts().sort_index()

# Display counts
print("Number of restaurants in each price range:")
print(price_range_counts)
```

```python
# Bar chart for price range
colors = ['red', 'green', 'yellow', 'blue']
plt.figure(figsize=(8,5))
plt.bar(price_range_counts.index.astype(str), price_range_counts.values, color=colors)
plt.title("The Distribution of Price Ranges Among The Restaurants")
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

#  Calculate The Percentage of Restaurants In Each Price Range Category.

```python
# Calculate percentages
total_restaurants = len(data)
price_range_percentages = (price_range_counts / total_restaurants) * 100

# Combine count and percentage into a DataFrame
result = pd.DataFrame({
    'Count': price_range_counts,
    'Percentage': price_range_percentages.round(2)
})

# Display result
print("Price range distribution with percentages:")
print(result)
```
