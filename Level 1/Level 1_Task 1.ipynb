# Task 1 : Top Cuisines

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
data
```

```python
# Display first few rows
data.head()
```

```python
# Extract 'Cuisines' Column and Clean Data

# Split multiple cuisines
all_cuisines = data['Cuisines'].dropna().str.split(', ')

# Flatten list
cuisine_list = [cuisine for sublist in all_cuisines for cuisine in sublist]

# Convert to pandas Series
cuisine_series = pd.Series(cuisine_list)
```

# Determine The Top Three Most Common Cuisines.

```python
# Count cuisine frequency
cuisine_counts = cuisine_series.value_counts()

# Top 3 cuisines
top_3_cuisines = cuisine_counts.head(3)

print("Top 3 Cuisines by Count:")
top_3_cuisines
```

```python
# Bar chart
colors = ['red', 'green', 'blue'] 
plt.figure(figsize=(8,5))
plt.bar(top_3_cuisines.index, top_3_cuisines.values, color=colors)
plt.title('The Top Three Most Common Cuisines', fontsize=14)
plt.ylabel('Count')
plt.xlabel('Cuisine')
plt.show()
```

#  Calculate The percentage of restaurants that serve each of the top cuisines.


```python
# Total number of cuisines
total_cuisine_entries = len(cuisine_series)

# Calculate percentages
top_3_percentages = (top_3_cuisines / total_cuisine_entries) * 100

# Combine both
result = pd.DataFrame({
    'Count': top_3_cuisines,
    'Percentage': top_3_percentages.round(2)
})

result
```

```python
# Pie chart
plt.figure(figsize=(6,6))
plt.pie(result['Percentage'], labels=result.index, autopct='%1.1f%%')
plt.title('The Percentage of Restaurants That Serve Each of The Top Cuisines', fontsize=14)
plt.show()
```
