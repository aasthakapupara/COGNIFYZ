# Task 1 : Restaurant Ratings

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
data
```

```python
# Display first few rows
data.head()
```

#  Analyze The Distribution of Aggregate Ratings And Determine The Most Common Rating Range.

```python
# Distribution of Aggregate ratings

# Create histogram
plt.figure(figsize=(8,5))
plt.hist(data['Aggregate rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Round ratings to nearest 0.5 for better grouping
data['Rounded Rating'] = data['Aggregate rating'].round(1)

# Count most common rating
rating_counts = data['Rounded Rating'].value_counts().sort_index(ascending=False)

# Display rating counts
rating_counts

# Most common rating value
most_common_rating = rating_counts.idxmax()
print(f"The most common rating is : {most_common_rating}")
```

#  Calculate The Average Number of Votes Received By Restaurants.

```python
# Calculate average number of votes
average_votes = data['Votes'].mean()

print(f"Average number of votes per restaurant : {average_votes:}")
```
