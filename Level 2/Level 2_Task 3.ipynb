# Task 3 : Geographic Analysis

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
data
```

```python
# Display first few rows
data.head()
```

```python
# Display columns to verify coordinate columns
data.columns
```

# Plot The Locations of Restaurants on A Map Using Longitude And Latitude Coordinates.

```python
# Plot Longitude vs Latitude
plt.figure(figsize=(10, 6))
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.5, s=10, color='skyblue', edgecolors='black')
plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

#  Identify Any Patterns or Clusters of Restaurants In Specific Areas. 

```python
# Simply display again — patterns can be visually observed
plt.figure(figsize=(10, 6))
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.5, s=10, color='orange', edgecolors='black')
plt.title('Possible Clusters of Restaurants')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
