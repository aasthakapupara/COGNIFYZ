# Task 4 : Restaurant Chains

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
data
```

```python
# Display first few rows
data.head()
```

# Identify If There Are Any Restaurant Chains Present In The Dataset.

```python
# Count occurrences of each restaurant name
restaurant_counts = data['Restaurant Name'].value_counts()

# Filter restaurant names that occur more than once (chains)
restaurant_chains = restaurant_counts[restaurant_counts > 1]

# Display the top 10 restaurant chains by number of locations
print("Top 10 Restaurant Chains by Number of Locations:")
print(restaurant_chains.head(10))
```

# Analyzing The Ratings and Popularity of Different Restaurant Chains.

```python
# Filter data to include only restaurants that are part of chains
chain_data = data[data['Restaurant Name'].isin(restaurant_chains.index)]

# Group by restaurant name, calculate average rating, total votes, and number of outlets
chain_analysis = chain_data.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'sum',
    'Restaurant Name': 'count'
}).rename(columns={'Restaurant Name': 'Number of Outlets'})

# Round values for cleaner display
chain_analysis['Aggregate rating'] = chain_analysis['Aggregate rating'].round(2)

# Sort by number of outlets (just for reference)
chain_analysis_sorted = chain_analysis.sort_values(by='Number of Outlets', ascending=False)
```

```python
# Sort by average rating
top_by_rating = chain_analysis.sort_values(by='Aggregate rating', ascending=False).head(10)

# Display top 10 highest-rated chains
print("\nTop 10 Highest-Rated Restaurant Chains:")
top_by_rating
```

```python
# Plot highest-rated restaurant chains

plt.figure(figsize=(12,6))
bars = plt.barh(top_by_rating.index[::-1], top_by_rating['Aggregate rating'][::-1], color='lightgreen')
plt.title('Top 10 Highest-Rated Restaurant Chains')
plt.xlabel('Average Rating')
plt.ylabel('Restaurant Chain')
plt.xlim(0, 5)
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Display values on bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.05, bar.get_y() + bar.get_height()/2, f'{width:.2f}', va='center')

plt.show()
```

```python
# Get the top 10 most popular chains based on total votes

# Sort by total votes
top_by_votes = chain_analysis.sort_values(by='Votes', ascending=False).head(10)

# Display top 10 most popular chains
top_by_votes
```

```python
# Plot most popular restaurant chains based on total votes

plt.figure(figsize=(13,6))
bars = plt.barh(top_by_votes.index[::-1], top_by_votes['Votes'][::-1], color='salmon')
plt.title('Top 10 Most Popular Restaurant Chains (Total Votes)')
plt.xlabel('Total Votes')
plt.ylabel('Restaurant Chain')
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Display values on bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 50, bar.get_y() + bar.get_height()/2, f'{int(width)}', va='center')

plt.show()
```
