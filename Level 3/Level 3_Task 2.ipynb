# Task 2 :  Votes Analysis

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
data
```

```python
# Display first few rows
data.head()
```

# Identify The Restaurants With The Highest And Lowest Number of Votes.

```python
# Restaurant with highest votes
max_votes = data.loc[data['Votes'].idxmax()]
print("Restaurant with highest votes:")
print(max_votes[['Restaurant Name', 'Votes', 'Aggregate rating']])

# Restaurant with lowest votes
min_votes = data.loc[data['Votes'].idxmin()]
print("\nRestaurant with lowest votes:")
print(min_votes[['Restaurant Name', 'Votes', 'Aggregate rating']])
```

# Analyze If There Is A Correlation Between The Number of Votes And The Rating of A Restaurant.

```python
# Calculate correlation coefficient
correlation = data['Votes'].corr(data['Aggregate rating'])
print(f"Correlation between Votes and Aggregate Rating: {correlation:.2f}")
```

```python
# Visualize the relationship using a scatter plot
plt.figure(figsize=(11,6))
plt.scatter(data['Votes'], data['Aggregate rating'], alpha=0.5, color='teal')
plt.title("Correlation Between Votes and Rating")
plt.xlabel("Votes")
plt.ylabel("Aggregate Rating")
plt.show()
```
