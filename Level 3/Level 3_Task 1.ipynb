```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
data
```

```python
# Display first few rows
data.head()
```

#  Analyze The Text Reviews To Identify The Most Common Positive And Negative Keywords.

```python
# Count occurrences of rating text categories
rating_counts = data["Rating text"].value_counts()
```

```python
# Visualize the distribution of rating text categories
plt.figure(figsize=(10,5))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.xticks(rotation=45)
plt.xlabel("Rating Category")
plt.ylabel("Count")
plt.title("Distribution of Rating Text Categories")
plt.show()
```

```python
# Display the most common rating text categories
rating_counts
```

```python
from collections import Counter
```

```python
# Define positive and negative ratings
positive_ratings = ["Excellent", "Very Good", "Good"]
negative_ratings = ["Average", "Poor"]
```

```python
# Count occurrences of positive and negative words
positive_words = " ".join(data[data["Rating text"].isin(positive_ratings)]["Rating text"]).lower().split()
negative_words = " ".join(data[data["Rating text"].isin(negative_ratings)]["Rating text"]).lower().split()
```

```python
# Get most common words
positive_word_counts = Counter(positive_words).most_common(5)
negative_word_counts = Counter(negative_words).most_common(5)

positive_word_counts, negative_word_counts
```

# Calculate The Average Length of Reviews And Explore If There Is A Relationship Between Review Length And Rating.

```python
# Calculate the length of rating text
data["Rating text length"] = data["Rating text"].apply(lambda x: len(x.split()))
```

```python
# Compute the average length of rating text
average_length = data["Rating text length"].mean()
```

```python
# Visualize the relationship between rating text length and aggregate rating
plt.figure(figsize=(10, 5))
sns.boxplot(x="Rating text", y="Rating text length", data=data, order=rating_counts.index)
plt.xticks(rotation=45)
plt.xlabel("Rating Category")
plt.ylabel("Rating Text Length")
plt.title("Relationship Between Rating Text Length and Rating Category")
plt.show()

average_length
```
